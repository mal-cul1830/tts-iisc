```python
def get_text(url):
  from bs4 import BeautifulSoup
  import requests
  page = requests.get(url)
  soup = BeautifulSoup(page.content, 'html.parser')
  p = soup.find_all('p')
  from bs4 import BeautifulSoup
  cleantext = BeautifulSoup(str(p), "lxml").text

  return cleantext
```

```python
!wget -O /tmp/lid.176.bin https://dl.fbaipublicfiles.com/fasttext/supervised-models/lid.176.bin
import fasttext
PRETRAINED_MODEL_PATH = '/tmp/lid.176.bin'
model = fasttext.load_model(PRETRAINED_MODEL_PATH)
```

```python
def isKannada(string):
    prediction = model.predict(string)
    if len(string) < 5 or len(string.split()) < 4:
      return False
    return prediction[0][0] == '__label__kn' and prediction[1][0] > 0.7
```

```python
for url in links:
  cleantext = get_text(url)
  fulllist = fulllist + list(filter(isKannada, re.split('\. |\n|\[|\]',cleantext)))
```
